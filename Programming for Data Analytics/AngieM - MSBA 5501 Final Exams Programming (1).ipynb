### Problem 2 

Write a function that takes a number as input and return the square root of the input.
Use exception handling below

try, except, else and finally 

to catch exception when the input is negative, in which case the result is None and
"Error: Negative Input" is printed.


```python
def mySqrt(x):
    # modify and add code here
    import math
    try:
        sqrt = math.sqrt(x)
    except:
        print("Error: Negative Input")
        sqrt = None
    else:
        sqrt = math.sqrt(x)
        print("result is", sqrt)
    finally:
        print("executing finally clause")
        return sqrt
```

If the index is less than 51 the segment is assigned to be "good"

If the index is 51 and higher but below 101, the segment is "moderate"

If the index is 101 and higher but below 151, the segment is "unhealthy for sensitives"

If the index is 151 and higher but below 201, the segment is "unhealthy"

If the index is 201 and higher but below 251, the segment is "very unhealthy"

If the index is 251 and higher, the segment is "hazardous"

return the segment result

2. Assign the segment of air index 43, 55, 75, 162 and 360
to get the result of "good", "moderate", "moderate", "unhealthy" and "hazardous"

3. Use list and iterate the list to calculate segment for a list of locations with air index 10, 53, 65, 78, 123, 135, 424, 262, 75, 88

```python
# code the function here
def airQuality(index):
    # write code here
# If the index is less than 51 the segment is assigned to be "good"
    if index < 51:
        quality = 'good'
# If the index is 51 and higher but below 101, the segment is "moderate"
    elif index < 101:
        quality = 'moderate'
# If the index is 101 and higher but below 151, the segment is "unhealthy for sensitives"
    elif index < 151:
        quality = 'unhealthy for sensitives'
# If the index is 151 and higher but below 201, the segment is "unhealthy"
    elif index < 201:
        quality = 'unhealthy'
# If the index is 201 and higher but below 251, the segment is "very unhealthy"
    elif index < 251:
        quality = 'very unhealthy'
# If the index is 251 and higher, the segment is "hazardous"
    else:
        quality = 'hazardous'
    return quality
```

### Problem 6

Write a function to calculate the Euclidean distance between two lists of any dimension.
The formula is given below:

$euclideanDistance([x_1, x_2,..., x_n], [y_1, y_2, ..., y_n]) = \sqrt{(x_1-y_1)^2 + (x_2-y_2)^2 + ... + (x_n-y_n)^2}$

Use assert exception if the lengths of the two input lists are not equal

```python
# Write a function to calculate the Euclidean distance between two lists of any dimension
# write code here
def euclideanDistance(X_i, X_j):
    import math
# Use assert exception if the lengths of the two input lists are not equal
    assert(len(X_i)==len(X_j)),"Error. The two lists need to have the same dimension"
    num = 0
    for x in range(len(X_i)):
        num += (X_i[x]-X_j[x])**2
    dist = math.sqrt(num)
    return dist
```

```python
import unittest
class EuclideanDisTest(unittest.TestCase):
    def testDist(self):
        X_1 = [-0.34556054, 0.53165302, -2.0899979,  -1.86463556, 0.49895049]
        X_2 = [-0.34556054, 0.53165302, -2.0899979,  -1.86463556, 0.49895049]
        self.assertAlmostEqual(euclideanDistance(X_1, X_2), 0.0, places=15)
        X_1 = [-0.70179092, -1.07561725, -0.88997364,  1.00354491,  0.74136834, 1.2346353]
        X_2 = [-0.34556054, 0.53165302, 2.0899979,  -1.86463556, 0.49895049, 3.2146352]
        self.assertAlmostEqual(euclideanDistance(X_1, X_2), 4.878121914236958, places=15)
    @unittest.expectedFailure
    def testDist2(self):
        X_1 = [0.12526706, 0.77269163, 0.51698169, 0.86463556, 1.1049951]
        X_2 = [-1.08149743,  2.31061441, -1.68924072,  1.86463556,  0.1049951, 0.4343434 ]
        self.assertAlmostEqual(euclideanDistance(X_1, X_2), 0.0, places=15)
```

two sentences or paragraphs.

You are asked to write a function for it.  The inputs are two lists of numbers indicating 
word counts in the dictionary based on some sentences such as "mary has a little lamb". 

Your function should successfully calculate the similarities of a few phrases in the given 
unit test.

$cosineSimilarity([x_1, x_2,..., x_n], [y_1, y_2, ..., y_n]) = \frac{x_1y_1 + x_2y_2 + ... + x_ny_n}{\sqrt{x_1^2 + x_2^2 + ... + x_n^2}\sqrt{y_1^2 + y_2^2 + ...+y_n^2}}$


```python
# write a function for it
# write code here
def cosineSimilarty(X_i, X_j):
    import numpy as np
    from numpy.linalg import norm
    a = X_i
    b = X_j
    cos_sim = np.dot(a,b)/(norm(a)*norm(b))
    return cos_sim
```
